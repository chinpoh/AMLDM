# DATA PREPROCESSING

```python
rep = {0: "Not_Survived", 1: "Survived"}
df.replace({'Survived' : rep}, inplace=True)
```

```python
df = df.drop(['PassengerId','Name','Parch','SibSp','Age','Ticket','Fare','Cabin'], axis=1)
df.head()
```

```python
dataset = []
for i in range(0,891):
    dataset.append([str(df.values[i,j]) for j in range(0, df.shape[1])])


oht = TransactionEncoder()
oht_ary = oht.fit(dataset).transform(dataset)
df1 = pd.DataFrame(oht_ary, columns=oht.columns_)
df1.head()
```

```python
oht_ary
```

```python
oht.columns_
```

# LEARNING MODE

```python
titanicsurvivor = apriori(df1, min_support=0.3, use_colnames=oht.columns_)
titanicsurvivor.head()
```

```python
my_rules = association_rules(titanicsurvivor, metric="lift", min_threshold=1)
```
