```python
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['Description'] = df['Description'].str.strip()
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')] 
df = df[~df['InvoiceNo'].str.contains('A')] 
df.head()
```

# DATA PREPROCESSING

```python
myshopcart = (df[df['Country'] =="United Kingdom"]
             .groupby(['InvoiceNo', 'Description'])['Quantity']
             .sum().unstack().reset_index().fillna(0)
             .set_index('InvoiceNo'))
```

```python
myshopcart.head()
```

```python
myshopcart.info()
```

```python
def my_encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

myshopcart_sets = myshopcart.applymap(my_encode_units)
myshopcart_sets.drop('POSTAGE', inplace=True, axis=1) 
```

# LEARNING MODE

```python
my_frequent_itemsets = apriori(myshopcart_sets, min_support=0.03, use_colnames=True)
```

```python
my_rules = association_rules(my_frequent_itemsets, metric="lift", min_threshold=1)
```
